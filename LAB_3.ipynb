```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
df = pd.read_csv('train_and_test2.csv')
```

```python
df.head(5)
```

```python
#checking the column names
for col in df.columns:
  print(col)
```

```python
df.isnull().sum()
```

**Task1:** Impute missing values in the Age column using the median.

```python
#there is no missing value in age column

median_age = df['Age'].median()


df['Age'].fillna(median_age, inplace=True)
```

**Task2: Fill missing values in the Embarked column with the most frequent value.**

```python
most_frequent_embarked = df['Embarked'].mode()[0]

most_frequent_embarked

df['Embarked'].fillna(most_frequent_embarked, inplace=True)
```

**Task3: Drop rows with missing Cabin data.**

```python
# df.dropna(df['Cabin'], inplace=True)

df.dropna()
```

```python
df.isnull().sum()
```

**Task4: Convert the Pclass column to a categorical variable.**

```python
print(df['Pclass'].dtype)
df['Pclass'] = df['Pclass'].astype('category')
print(df['Pclass'].dtype)
```

**Task5: Ensure the Fare column is a float, not an integer.**

```python
print(df['Fare'].dtype)
df['Fare'] = df['Fare'].astype(float)
print(df['Fare'].dtype)
```

**Task6: Analyze if Age has outliers and handle them accordingly.**

```python
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1


lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


outliers = df[(df['Age'] < lower_bound) | (df['Age'] > upper_bound)]
print(outliers)

no_outliers = df[(df['Age'] >= lower_bound) & (df['Age'] <= upper_bound)]
# any_outlier = df[(df['Age'] < lower_bound) | (df['Age'] > upper_bound)]
# print("#####################################################################3")
# print(any_outlier)
```

**Task7: Remove any duplicate rows if found.**

```python
df = df.drop_duplicates()
```

```python
x = df.drop('2urvived', axis=1)
y = df['2urvived']


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)


lr = LogisticRegression(max_iter=200)
lr.fit(x_train, y_train)


y_pred = lr.predict(x_test)


accuracy = metrics.accuracy_score(y_test, y_pred) * 100
print("Accuracy:", accuracy)
```
